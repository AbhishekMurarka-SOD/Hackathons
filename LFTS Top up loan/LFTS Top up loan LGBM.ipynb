```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv("train_Data.csv")
df.head(10)
```

```python
df.drop(['Area','City','MaturityDAte','AuthDate','DisbursalDate'],axis=1,inplace=True)
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
df.drop(['AmountFinance'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
cat_col_mask=df.dtypes==object
cat_col=df.columns[cat_col_mask].tolist()
le = LabelEncoder()
df[cat_col] = df[cat_col].apply(lambda col: le.fit_transform(col))
df[cat_col].head(10)
```

```python
df.info()
```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df['Top-up Month']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from lightgbm import LGBMClassifier
from sklearn import metrics
accuracy = []
precision = []
recall = []
f1_score = []

def calculate_metrics(y_test,y_pred):
    acc = metrics.accuracy_score(y_true = y_test, y_pred = y_pred)
    precision = metrics.precision_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    recall = metrics.recall_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    f1 = metrics.f1_score(y_true = y_test, y_pred = y_pred,pos_label='positive',average='macro')
    
    accuracy.append(acc)
    #precision.append(precision)
    #recall.append(recall)
    f1_score.append(f1)
    
kfold_mean=[]
kfold_max=[]

def calculate_kfold(estimator):
    accuracies = cross_val_score(estimator,x,y, cv=20)
    kfold_mean.append(accuracies.mean())
    kfold_max.append(accuracies.max())
```

```python
lgbm=LGBMClassifier()
lgbm.fit(x_train,y_train)
lgbmpred=lgbm.predict(x_test)
calculate_metrics(y_test, lgbmpred)
calculate_kfold(lgbm)
```

```python
print(f1_score)
```

```python
from sklearn.model_selection import RandomizedSearchCV
params = {'max_depth' : [1,5,9,15],
          'num_leaves': [24,64,84,100],
          'learning_rate': [0.05,0.2,0.1,0.3],
          'n_estimators': [100,200,300,500],
         }
hypelib=RandomizedSearchCV(LGBMClassifier(),params, scoring='f1_macro')
hypelib.fit(x_train,y_train)
```

```python
hypelib.best_params_, hypelib.best_score_
```
